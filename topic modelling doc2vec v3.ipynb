```python
def remove_stopwords(s):
    hasil_replace = s.replace("'", '').replace("&", '').replace("{", '').replace("}", '').replace("-", '').replace("/", '').replace("!", '').replace(",",'').replace('''"''', '').replace('.', '').replace('(', '').replace(')', '').replace(':', '').replace('?', '').replace('`', '').replace(';', '')
    f = open('stopwordbahasa.csv', 'r')

    stopwords = []
    for lines in f:
        lines = lines.strip('\n')
        stopwords.append(lines)


    corpus_stopwrods = hasil_replace.split(' ')
    hasil_stopword = []

    for x in corpus_stopwrods:
        if x not in stopwords:
            hasil_stopword.append(x)


        
    hasil = " ".join(hasil_stopword)
    hasil = hasil.replace('\n', '')


    
    
    return(hasil)
```

```python
def create_tagged_document(list_of_list_of_words, nama_file):
    for i, list_of_words in enumerate(list_of_list_of_words):
        a = nama_file[i]
        yield gensim.models.doc2vec.TaggedDocument(list_of_words, [a])

documents = list(create_tagged_document(my_corpus, nama_files))
```
